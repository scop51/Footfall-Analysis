```python
import pandas as pd

FOOTFALL_DATA = r"C:\Users\sathy\python\weather data\processed_data\final_eda_nyc_weather_2024.csv"
df = pd.read_csv(FOOTFALL_DATA)
df.head()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
from scipy.stats import zscore, pearsonr, spearmanr, ttest_ind
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error
from statsmodels.tsa.arima.model import ARIMA
from prophet import Prophet
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from sklearn.preprocessing import StandardScaler, PolynomialFeatures
from sklearn.model_selection import GridSearchCV
```

```python
# Convert date column to datetime format
df["date"] = pd.to_datetime(df["date"])
df.set_index("date", inplace=True)
```

```python
# Encode 'season' as numerical categories
df["season_encoded"] = df["season"].astype("category").cat.codes
```

```python
# Compute Pearson and Spearman correlations
pearson_corr = df[["mall_footfall", "atm_outlet_footfall", "season_encoded", "holiday"]].corr(method="pearson")
spearman_corr = df[["mall_footfall", "atm_outlet_footfall", "season_encoded", "holiday"]].corr(method="spearman")
```

```python
# Perform Hypothesis Testing
pearson_pval = pearsonr(df["mall_footfall"], df["holiday"])[1]
spearman_pval = spearmanr(df["mall_footfall"], df["holiday"])[1]

print("Pearson Correlation P-Value:", pearson_pval)
print("Spearman Correlation P-Value:", spearman_pval)
```

```python
# Feature Engineering: Add interaction and polynomial features
poly = PolynomialFeatures(degree=2, include_bias=False)
features = ["temperature_max", "temperature_min", "precipitation", "holiday", "season_encoded"]
X_poly = poly.fit_transform(df[features])
X_poly = pd.DataFrame(X_poly, columns=poly.get_feature_names_out(features))

# Scaling features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_poly)
X_scaled = pd.DataFrame(X_scaled, columns=X_poly.columns)
```

```python
# Target Variables
y_mall = df["mall_footfall"]
y_atm = df["atm_outlet_footfall"]

# Train-Test Split
X_train, X_test, y_train_mall, y_test_mall = train_test_split(X_scaled, y_mall, test_size=0.2, random_state=42)
X_train, X_test, y_train_atm, y_test_atm = train_test_split(X_scaled, y_atm, test_size=0.2, random_state=42)
```

```python
# Linear Regression
lr_model_mall = LinearRegression()
lr_model_mall.fit(X_train, y_train_mall)
lr_preds_mall = lr_model_mall.predict(X_test)
print("Linear Regression MAE (Mall Footfall):", mean_absolute_error(y_test_mall, lr_preds_mall))

lr_model_atm = LinearRegression()
lr_model_atm.fit(X_train, y_train_atm)
lr_preds_atm = lr_model_atm.predict(X_test)
print("Linear Regression MAE (ATM Outlet Footfall):", mean_absolute_error(y_test_atm, lr_preds_atm))
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))

# Scatter plot for Mall Footfall
plt.subplot(1, 2, 1)
plt.scatter(y_test_mall, lr_preds_mall, alpha=0.5, label="Mall Footfall")
plt.plot([min(y_test_mall), max(y_test_mall)], [min(y_test_mall), max(y_test_mall)], color='red', linestyle='--')  
plt.xlabel("Actual Values")
plt.ylabel("Predicted Values")
plt.title("Linear Regression: Predicted vs Actual (Mall Footfall)")
# plt.legend()
# plt.show()


# Scatter plot for ATM Outlet Footfall
plt.subplot(1, 2, 2)
plt.scatter(y_test_atm, lr_preds_atm, alpha=0.5, label="ATM Outlet Footfall", color='green')
plt.plot([min(y_test_atm), max(y_test_atm)], [min(y_test_atm), max(y_test_atm)], color='red', linestyle='--')  
plt.xlabel("Actual Values")
plt.ylabel("Predicted Values")
plt.title("Linear Regression: Predicted vs Actual (ATM Outlet Footfall)")

plt.tight_layout()
plt.show()
```

```python
# Random Forest Regression
rf_model_mall = RandomForestRegressor(n_estimators=100, max_depth=10, random_state=42)
rf_model_mall.fit(X_train, y_train_mall)
rf_preds_mall = rf_model_mall.predict(X_test)
print("Random Forest MAE (Mall Footfall):", mean_absolute_error(y_test_mall, rf_preds_mall))

rf_model_atm = RandomForestRegressor(n_estimators=100, max_depth=10, random_state=42)
rf_model_atm.fit(X_train, y_train_atm)
rf_preds_atm = rf_model_atm.predict(X_test)
print("Random Forest MAE (ATM Outlet Footfall):", mean_absolute_error(y_test_atm, rf_preds_atm))
```

```python
# Plot for Mall Footfall
plt.figure(figsize=(12, 6))

# Actual vs Predicted for Mall Footfall
plt.subplot(1, 2, 1)
plt.scatter(y_test_mall, rf_preds_mall, alpha=0.5)
plt.plot([min(y_test_mall), max(y_test_mall)], [min(y_test_mall), max(y_test_mall)], color='red', linestyle='--')
plt.xlabel('Actual Mall Footfall')
plt.ylabel('Predicted Mall Footfall')
plt.title('Random Forest Regression: Mall Footfall')

# Plot for ATM Outlet Footfall
plt.subplot(1, 2, 2)
plt.scatter(y_test_atm, rf_preds_atm, alpha=0.5)
plt.plot([min(y_test_atm), max(y_test_atm)], [min(y_test_atm), max(y_test_atm)], color='red', linestyle='--')
plt.xlabel('Actual ATM Outlet Footfall')
plt.ylabel('Predicted ATM Outlet Footfall')
plt.title('Random Forest Regression: ATM Outlet Footfall')

plt.tight_layout()
plt.show()
```

```python
# Time-Series Forecasting
# ARIMA for mall footfall
arima_model_mall = ARIMA(df["mall_footfall"], order=(5,1,0))
arima_result_mall = arima_model_mall.fit()
print(arima_result_mall.summary())

# ARIMA for ATM footfall
arima_model_atm = ARIMA(df["atm_outlet_footfall"], order=(5,1,0))
arima_result_atm = arima_model_atm.fit()
print(arima_result_atm.summary())

# Prophet for mall footfall
df_prophet_mall = df.reset_index()[["date", "mall_footfall"]].rename(columns={"date": "ds", "mall_footfall": "y"})
prophet_model_mall = Prophet()
prophet_model_mall.fit(df_prophet_mall)
future_mall = prophet_model_mall.make_future_dataframe(periods=30)
forecast_mall = prophet_model_mall.predict(future_mall)
prophet_model_mall.plot(forecast_mall)
plt.show()

# Prophet for ATM footfall
df_prophet_atm = df.reset_index()[["date", "atm_outlet_footfall"]].rename(columns={"date": "ds", "atm_outlet_footfall": "y"})
prophet_model_atm = Prophet()
prophet_model_atm.fit(df_prophet_atm)
future_atm = prophet_model_atm.make_future_dataframe(periods=30)
forecast_atm = prophet_model_atm.predict(future_atm)
prophet_model_atm.plot(forecast_atm)
plt.show()
```
